```python
import pandas as pd
import numpy as np
import statsmodels.stats.proportion as sp

# 讀取資料 (從 notebooks 往上一層進入 data/raw)
df = pd.read_csv('../data/raw/YRBS_2007.csv')

# Task 2: 定義變數
# Group: WhatIsYourSex / Response: CurrentCigaretteUse
df_study = df[['WhatIsYourSex', 'CurrentCigaretteUse']].copy()
```

```python
# Task 3: 重編碼規則
# 原始碼 2-7 代表過去 30 天有吸菸 (Success = 1)
# 原始碼 1 代表沒吸菸 (Failure = 0)
def recode_smoking(x):
    if x == 1:
        return 0
    elif 2 <= x <= 7:
        return 1
    else:
        return np.nan

df_study['smoke_binary'] = df_study['CurrentCigaretteUse'].apply(recode_smoking)

# 排除缺失值 (Task 16: 避免忽略缺失值的錯誤 [cite: 1634])
df_clean = df_study.dropna()
```

```python
import numpy as np
import statsmodels.stats.proportion as sp

# 1. 確認資料是否讀入 (任務 11: 顯示前五筆)
print("--- 資料預覽 ---")
print(df.head())

# 2. 資料清理與重編碼 (任務 3)
# 根據手冊規則: 2-7 代表有吸菸 (1), 1 代表沒吸菸 (0) 
def recode_smoking(x):
    if x == 1:
        return 0
    elif 2 <= x <= 7:
        return 1
    else:
        return np.nan

df['smoke_binary'] = df['CurrentCigaretteUse'].apply(recode_smoking)
df_clean = df.dropna(subset=['WhatIsYourSex', 'smoke_binary'])

# 3. 製作摘要表 (任務 8 & 11) 
# 注意: 請確認資料中 1=Male, 2=Female
summary = df_clean.groupby('WhatIsYourSex')['smoke_binary'].agg(['sum', 'count'])
summary.columns = ['Smokers', 'Total_N']
summary['Proportion'] = summary['Smokers'] / summary['Total_N']
print("\n--- 摘要統計表 ---")
print(summary)

# 4. 兩樣本比例 Z 檢定 (任務 4 & 6) 
counts = summary['Smokers'].values
nobs = summary['Total_N'].values
z_stat, p_val = sp.proportions_ztest(counts, nobs)

print(f"\n--- 統計檢定結果 ---")
print(f"Z 統計量: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("結論: 在 alpha=0.05 下，男女生的吸菸比例有顯著差異。")
else:
    print("結論: 在 alpha=0.05 下，男女生的吸菸比例沒有顯著差異。")
```

```python
import numpy as np
import statsmodels.stats.proportion as sp

# 1. 資料清理與重編碼 (任務 3)
# 根據手冊 Q1 規則: 1 代表沒吸菸 (0), 2-7 代表有吸菸 (1)
def recode_smoking(x):
    if x == 1:
        return 0
    elif 2 <= x <= 7:
        return 1
    else:
        return np.nan

# 執行重編碼
df['smoke_binary'] = df['CurrentCigaretteUse'].apply(recode_smoking)

# 排除缺失值 (確保統計準確)
df_clean = df.dropna(subset=['WhatIsYourSex', 'smoke_binary'])

# 2. 製作摘要統計表 (任務 8 & 11)
# 1=Male, 2=Female (請對照手冊定義)
summary = df_clean.groupby('WhatIsYourSex')['smoke_binary'].agg(['sum', 'count'])
summary.columns = ['Smokers', 'Total_N']
summary['Proportion'] = summary['Smokers'] / summary['Total_N']

print("--- 摘要統計表 (Task 8/11) ---")
print(summary)

# 3. 兩樣本比例 Z 檢定 (任務 4 & 6)
counts = summary['Smokers'].values
nobs = summary['Total_N'].values
z_stat, p_val = sp.proportions_ztest(counts, nobs)

print(f"\n--- 統計檢定結果 (Task 6) ---")
print(f"Z 統計量 (Z-statistic): {z_stat:.4f}")
print(f"P-value: {p_val:.4e}") # 使用科學符號，因為通常會很小

# 4. 簡單結論
if p_val < 0.05:
    print("\n結論: P-value 小於 0.05，男女吸菸比例有顯著差異。")
else:
    print("\n結論: P-value 大於 0.05，男女吸菸比例沒有顯著差異。")
```
