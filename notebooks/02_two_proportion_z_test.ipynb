```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest, confint_proportions_2indep

# Load data
df = pd.read_csv('../data/processed/yrbs_cleaned_and_recoded.csv')
```

```python
# TWO-PROPORTION Z TEST

# Split data into two groups
male = df[df["WhatIsYourSex"] == 1]
female = df[df["WhatIsYourSex"] == 0]

# proportions

# Calculate smoking proportions (mean of binary variable)
male_yes = male["CurrentCigaretteUse"].mean()
female_yes = female["CurrentCigaretteUse"].mean()

# Compute non-smoking proportions
male_no = 1 - male_yes
female_no = 1 - female_yes

# Print variable definitions
print("1.Variable selection:")
print("  Grouping variable: WhatIsYourSex (Male= 1 vs Female= 0)")
print("  Response variable: CurrentCigaretteUse (No= 0,Yes= 1)")

print("\n2.Smoking Proportions:")

# Male results
print("  Male:")
print(f"    Smoke (Yes): {male_yes:.4f}")
print(f"    Not smoke (No): {male_no:.4f}")

# Female results
print("  Female:")
print(f"    Smoke (Yes): {female_yes:.4f}")
print(f"    Not smoke (No): {female_no:.4f}")

# z-test

# Count number of smokers in each group
count = np.array([
    male["CurrentCigaretteUse"].sum(),
    female["CurrentCigaretteUse"].sum()
])

# Sample sizes of each group
nobs = np.array([
    len(male),
    len(female)
])

# Run two-proportion z-test
z_stat, p_value = proportions_ztest(count, nobs)

# Print test results
print("\n3.Z-test result:")
print("  Z-statistic:", z_stat)
print("  P-value:", p_value)
```
