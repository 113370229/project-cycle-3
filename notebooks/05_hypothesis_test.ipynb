```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest, confint_proportions_2indep

# Load data
df = pd.read_csv('../data/processed/yrbs_cleaned_and_recoded.csv')
```

```python
# Split data into two groups based on gender
male = df[df["WhatIsYourSex"] == 1]
female = df[df["WhatIsYourSex"] == 0]

# Count number of smokers (sum of 1s) in each group
count = np.array([
    male["CurrentCigaretteUse"].sum(),
    female["CurrentCigaretteUse"].sum()
])

# Total sample size in each group
nobs = np.array([
    len(male),
    len(female)
])

# Perform two-proportion z-test
z_stat, p_value = proportions_ztest(count, nobs)

# Hypothesis Test & Conclusion

alpha = 0.05  # significance level

print("1.Hypothesis Test:")
print("  H0: p_male = p_female")
print("  H1: p_male ≠ p_female\n")

# Display test results
print("2.Hypothesis Test Result: ")
print("  Z-statistic:", z_stat)
print("  P-value:", p_value)

# Make decision based on p-value
if p_value < alpha:
    print("\n3.Decision: Reject H0")
    print("\n4.Conclusion: There is sufficient statistical evidence at the 0.05 significance level to conclude that the proportion of current cigarette use is different between male and female students.")
else:
    print("\n3.Decision: Fail to reject H0")
    print("\n4.Conclusion: There is not sufficient statistical evidence at the 0.05 significance level to conclude that the proportion of current cigarette use is different between male and female students.")
```
